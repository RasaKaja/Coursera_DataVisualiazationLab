## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import matplotlib.pyplot as plt
```

```python
QUERY = """
SELECT ConvertedComp FROM master"""
df_converted = pd.read_sql_query(QUERY,conn)
# df_converted.head()
```

```python
plt.figure(figsize=(10, 6))
df_converted['ConvertedComp'].plot(kind = 'hist', bins = 50)
plt.title('Distribution of ConvertedComp')
plt.xlabel('ConvertedComp (USD)')
plt.ylabel('Frequency')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
import seaborn as sns

QUERY = """
SELECT Age FROM master
"""
df_age = pd.read_sql_query(QUERY,conn)
# df_age.head()

plt.figure(figsize=(10, 6))
sns.boxplot(x=df_age['Age'], color='skyblue')
plt.title('Box Plot of Age')
plt.xlabel('Age')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
QUERY = """
SELECT Age, WorkWeekHrs FROM master
"""
df_scatter = pd.read_sql_query(QUERY,conn)
# df_scatter.head()

sns.scatterplot(x="Age", y = "WorkWeekHrs", data = df_scatter)
plt.title('Scatter Plot of Age and WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('Work Week Hours')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs FROM master
"""
df_bubble = pd.read_sql_query(QUERY,conn)
# df_bubble.head()

plt.figure(figsize=(10,6))
sns.scatterplot(x = 'WorkWeekHrs', y = 'CodeRevHrs', size = 'Age', alpha = 0.5, data = df_bubble)

plt.title('Bubble Plot of CodeRevHrs and WorkWeekHrs')
plt.xlabel('Work Week Hrs')
plt.ylabel('Code Review Hours')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
QUERY = """
SELECT DISTINCT DatabaseDesireNextYear as Language, count(Respondent) as Total 
FROM DatabaseDesireNextYear 
group by Language 
order by Total desc
LIMIT 5

"""
df_language_next = pd.read_sql_query(QUERY,conn)
# df_language_next.head()

pie_labels = df_language_next['Language'] 
pie_sizes = df_language_next['Total']

fig, ax = plt.subplots()
ax.pie(pie_sizes, labels = pie_labels, autopct='%1.1f%%')

plt.title('Pie Chart of TOP 5 databases that respondents whant to learn next')

plt.show()
```

```python
QUERY = """
SELECT DISTINCT LanguageDesireNextYear as Language, count(Respondent) as Total 
FROM LanguageDesireNextYear 
group by Language 
order by Total desc
LIMIT 5

"""

df_language_next = pd.read_sql_query(QUERY,conn)
print(df_language_next)

pie_labels = df_language_next['Language'] 
pie_sizes = df_language_next['Total']

fig, ax = plt.subplots()
ax.pie(pie_sizes, labels = pie_labels, autopct='%1.1f%%')

plt.title('Pie Chart of TOP 5 databases that respondents whant to learn next')

plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs 
FROM master
Where Age >= 30 and Age <=35
"""
# df_stacked.head()

df_stacked = pd.read_sql_query(QUERY,conn)

df_stacked_median = df_stacked.groupby('Age')[['WorkWeekHrs', 'CodeRevHrs']].median()
# print(df_stacked_median)

# Plot a stacked bar chart
ax = df_stacked_median.plot.bar(stacked=True, figsize=(10, 6), title='Hours of Work Week and Code Review')

# Add labels and legend
plt.xlabel('Age')
plt.ylabel('Median Hours')
plt.legend(title='Activity')

# Show the plot
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
QUERY = """
SELECT ConvertedComp, Age
FROM master
Where Age >= 45 and Age <=60
"""
df_lineCart = pd.read_sql_query(QUERY,conn)
# df_lineCart.head()

df_lineCart_median = df_lineCart.groupby('Age')[['ConvertedComp']].median()
# print(df_lineCart_median)

df_lineCart_median.plot.line()
```
